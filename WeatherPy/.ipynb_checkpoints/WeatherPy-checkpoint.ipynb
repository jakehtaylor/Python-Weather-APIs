```python
 # Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
import time
from scipy.stats import linregress, pearsonr
from citipy import citipy
from config import w_api
```

```python
url = 'http://api.openweathermap.org/data/2.5/weather?q='

latitudes = np.random.uniform(-90, 90, size=1500)
longitudes = np.random.uniform(-180, 180, size=1500)
coords = zip(latitudes, longitudes)
cities = []

for ll in coords:
    if citipy.nearest_city(ll[0], ll[1]).city_name not in cities:
        cities.append(citipy.nearest_city(ll[0], ll[1]).city_name)

len(cities)
```

```python
lat = []
lon = []
temp = []
humid = []
wind = []
clouds = []
country = []
name = []

master = [lat, lon, temp, humid, wind, clouds, country, name]

inds = {'coord' : ['lat', 'lon'],
        'main' : ['temp', 'humidity'],
        'wind' : ['speed'],
        'clouds' : ['all'],
        'sys' : ['country'],
        'name' : None,
       }

# setup: I have an empty list for each datapoint I want to gather, as well as a list of those lists. the 'inds'
# dict contains the indices of each of those datapoints in the open weather api response.

j = 1
# j is the number of the request
f = 0
# f is the number of cities we have found data for

for city in cities:
    
    m = 0
    # m is the index for the master list
    
    weather = requests.get(url + city + '&units=imperial' + f'&appid={w_api}').json()
    
    if 'message' in weather.keys():
        print(weather['message'] +': '+ city)
        j+= 1
        # the api only returns a message value if something went wrong, so we can use this 
        # to display the 'city not found' message and avoid index errors.
        continue
    
    for i in inds:
        if inds[i]:
            for sub in inds[i]:
                master[m].append(weather[i][sub])
                m += 1
        else:
            master[m].append(weather[i])
            m += 1
            
        # this loop cycles through the master list and the combination of indices and sub-indices 
        # associated with each data category in the master list, and appends the found data to the 
        # category. 'name' doesn't have a sub-index, hence the if/else. 
            
    print(f'gathering data for {city}... (request #{j} out of {len(cities)})')
    j += 1
    f +=1

    if f == 1:
        dt = time.gmtime(weather['dt'])
        date = f'({dt.tm_mon}/{dt.tm_mday}/{dt.tm_year})'
        
        # for the sake of efficency, I only get the date from the first city we find data for,
        # as we will only be using it to label the graphs.

print(f'Done. Found data for {f} out of {len(cities)} cities.')
```

```python
w_data = pd.DataFrame({
    'City' : master[7],
    'Latitude': master[0],
    'Longitude' : master[1],
    'Temperature' : master[2],
    'Humidity (%)' : master[3],
    'Wind Speed (mph)' : master[4],
    'Cloudiness (%)' : master[5],
    'Country' : master[6],
})

w_data = w_data[w_data['Humidity (%)'] < 101]
```

```python
w_data.to_csv('output_data.csv', index=False)
w_data
```

```python
plt.scatter(w_data['Latitude'], w_data['Temperature'])
plt.title(f'City Latitude vs Temperature {date}')
plt.xlabel('Latitude')
plt.ylabel('Temperature (F)')
plt.savefig('images/lat_temp.png')
```

From this scatterplot we can see a fairly strong relationship between temperature and latitude, with temperatures increasing as we approach the equator. Interestingly, though, it appears the peak of the curve aligns with latitude = 20° rather than the exact equator. This might be a result of the earth's tilt and rotation at this time of year, or indeed the exact day that this data was gathered.

```python
plt.scatter(w_data['Latitude'], w_data['Humidity (%)'])
plt.title(f'City Latitude vs Humidity {date}')
plt.xlabel('Latitude')
plt.ylabel('Humidity (%)')
plt.savefig('images/lat_hum.png')
```

There are a few different trends we can pick out here, modeled by a w-shaped polynomial. First, we can see a general quadratic shape of higher humidities between -20° and 20°. This is the Earth's tropical zone, containing all of the world's major rainforests, and characterized by higher temperatures and humidity. The tropics are Flanked (20° to 50° and -20° to -50°) by desert climates, representing the lowest humidities on Earth. We can see that the numbers start to rise again once we leave the desert zones and move towards the poles.

```python
plt.scatter(w_data['Latitude'], w_data['Wind Speed (mph)'])
plt.title(f'City Latitude vs Wind Speed {date}')
plt.xlabel('Latitude')
plt.ylabel('Wind Speed (mph)')
plt.savefig('images/lat_win.png')
```

There don't appear to be any strong, observable trends represented by this graph, indicating that there isn't a relationship between latitude and wind speed. We can see, though, that the majority of the world experienced wind speeds below 15 mph on this day.

```python
plt.scatter(w_data['Latitude'], w_data['Cloudiness (%)'])
plt.title(f'City Latitude vs Cloudiness {date}')
plt.xlabel('Latitude')
plt.ylabel('Cloudiness (%)')
plt.savefig('images/lat_cld.png')
```

Again, there doesn't seem to be any relationship here. The horizontal lines forming around particular percentages are interesting, but I would guess that would have to do with the way the percentages are calculated and given to us by the api, rather than being an actual phenomenon. 

```python
north = w_data[w_data['Latitude'] > 0]
north.name = 'North'
south = w_data[w_data['Latitude'] < 0]
south.name = 'South'
```

```python
def regress_plot(df, x, y):
    
    plt.scatter(df[x], df[y])
    plt.title(f'{df.name}ern Hemisphere {x} vs {y} {date}')
    plt.xlabel(x)
    plt.ylabel(y)
    
    (slope, intercept, rvalue, pvalue, stderr) = linregress(df[x], df[y])
    reg = df[x] * slope + intercept
    plt.plot(df[x], reg, 'r')
    plt.savefig(f'images/{df.name}_{x[0:3]}_{y[0:3]}.png')
    return f"Pearson's R : {pearsonr(df[x], df[y])[0]}"
    
    
regress_plot(north, 'Latitude', 'Temperature')
```

```python
regress_plot(south, 'Latitude', 'Temperature')
```

Here we can see that splitting the original latitude vs. temperature graph into the northern and southern hemispheres gives us two high Pearson's R scores, indicating strong correlations, one negative (north) and one positive (south), though both are negative if we consider distance from the equator instead of negative and positive latitude.

```python
regress_plot(north, 'Latitude', 'Humidity (%)')
```

```python
regress_plot(south, 'Latitude', 'Humidity (%)')
```

As we stated in our analysis for the combined graph, there is no linear correlation when we examine the hemispheres as a whole, confirmed by the low Pearson's R scores. However, if we split these graphs again on the edge of the tropics, we would see 4 linear correlations, as explained in the original analysis.

```python
regress_plot(north, 'Latitude', 'Wind Speed (mph)')
```

```python
regress_plot(south, 'Latitude', 'Wind Speed (mph)')
```

Just as in the combined graph, as proven by the Pearson's R scores, there is no correlation to be found between latitude and wind speed for either hemisphere.

```python
regress_plot(north, 'Latitude', 'Cloudiness (%)')
```

```python
regress_plot(south, 'Latitude', 'Cloudiness (%)')
```

Again, through the Pearson's R scores we can see that there is no correlation between latitude and cloudiness for either hemisphere.
